```python
import gridgeo


gridgeo.__version__
```

```python
url = 'http://geoport.whoi.edu/thredds/dodsC/clay/usgs/users/jcwarner/Sandy/triple_nest/00_dir_NYB07.ncml'

grid = gridgeo.GridGeo(
    url,
    standard_name='sea_water_potential_temperature'
)
```

```python
grid.mesh
```

```python
print(f'The grid has {len(grid.polygons)} polygons, showing the first 5.')

grid.geometry[:5]
```

```python
grid.outline
```

```python
%matplotlib inline
import cartopy.crs as ccrs
import matplotlib.pyplot as plt


fig, ax = plt.subplots(
    figsize=(9, 9),
    subplot_kw={'projection': ccrs.PlateCarree()}
)

ax.plot(
    grid.x, grid.y, 'darkgray',
    grid.x.T, grid.y.T, 'darkgray',
    alpha=0.25,
)

ax.coastlines(resolution='10m');
```

```python
import folium


x, y = grid.outline.centroid.xy

m = folium.Map(location=[y[0], x[0]])
folium.GeoJson(grid.outline.__geo_interface__).add_to(m)
m.fit_bounds(m.get_bounds())

m
```
